# Post pruning decision trees with cost complexity pruning
.. currentmodule:: sklearn.tree

#### The :class:DecisionTreeClassifier provides parameters such as min_samples_leaf and max_depth to prevent a tree from overfiting. Cost complexity pruning provides another option to control the size of a tree. In :class:DecisionTreeClassifier, this pruning technique is parameterized by the cost complexity parameter, ccp_alpha. Greater values of ccp_alpha increase the number of nodes pruned. Here we only show the effect of ccp_alpha on regularizing the trees and how to choose a ccp_alpha based on validation scores.

### See also minimal_cost_complexity_pruning for details on pruning.

```python
#to overcome the overfitting in decision tree we use postpruning
#we are going to use ccp_alpha,we if provide some value to ccp_alpha,this will restrict the max depth to some levels
import sklearn
sklearn.__version__
```

```python
%matplotlib inline
```

```python
print(__doc__)
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_breast_cancer
from sklearn.tree import DecisionTreeClassifier
```

```python
X, y = load_breast_cancer(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)

clf = DecisionTreeClassifier(random_state=0)
clf.fit(X_train,y_train)
```

```python
pred=clf.predict(X_test)
from sklearn.metrics import accuracy_score
accuracy_score(y_test, pred)
```

```python
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(clf,filled=True)
```

```python
path=clf.cost_complexity_pruning_path(X_train,y_train) #this will give us two values alpha and impurities
ccp_alphas,impurities=path.ccp_alphas,path.impurities  #ccp_alphas  will find the point with respect to the leaf nodes
```

```python
ccp_alphas
```

```python
clfs = []
for ccp_alpha in ccp_alphas:
    clf = DecisionTreeClassifier(random_state=0, ccp_alpha=ccp_alpha)
    clf.fit(X_train, y_train)
    clfs.append(clf)
print("Number of nodes in the last tree is: {} with ccp_alpha: {}".format(
      clfs[-1].tree_.node_count, ccp_alphas[-1]))
```

```python
train_scores = [clf.score(X_train, y_train) for clf in clfs]
test_scores = [clf.score(X_test, y_test) for clf in clfs]

fig, ax = plt.subplots()
ax.set_xlabel("alpha")
ax.set_ylabel("accuracy")
ax.set_title("Accuracy vs alpha for training and testing sets")
ax.plot(ccp_alphas, train_scores, marker='o', label="train",
        drawstyle="steps-post")
ax.plot(ccp_alphas, test_scores, marker='o', label="test",
        drawstyle="steps-post")
ax.legend()
plt.show() #at alpha value at 0.012 we have a ggod accuracy of test and train
```

```python
clf=DecisionTreeClassifier(random_state=0,ccp_alpha=0.012)
```

```python
clf.fit(X_train,y_train)
```

```python
predict=clf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,predict)
```
